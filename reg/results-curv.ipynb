```python
plt.close()
newdf = pd.concat([df['Fracture surface area total - m2'], df['Vertical constrast of minimum horizontal stress'],df['Horizontal stress Anisotropy (Shmax / Shmin) ']], axis=1, keys=['y', 'x1','x2'])



model = ols("y ~ I(x1**2) + I(x2**2) + x1:x2 + x1 + x2", newdf).fit()
model.summary()
```

```python
model.params[1]

def quadpoly(x,y,m):
    # takes in an x and y value and a fitted model and returns the z resultant value
    c = m.params
    
    val = c[1]*(x**2) + c[2]*(y**2) + c[3]*(x*y) + c[4]*x + c[5]*y + c[0]
    
    return (val)
    
```
